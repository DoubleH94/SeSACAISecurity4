```python
# 데이터프레임 : 결측치
# 결측치는 데이터 입력 과정에서 누락, 데이터 수집 장치의 오류/손상, 서로 다른 데이터 유형을 통합하는 과정에서 발생함
# ex) np.nan, None, 공백, 없음
import pandas as pd
import numpy as np

df5 = pd.DataFrame({
    'name': ['Alice','Bob','Charlie','David'],
    'age': [25,np.nan,30,22],
    'city': ['Seoul','Busan',None,'Daegu']
})

print(df5)
```

```python
# 데이터 프레임 : 결측치 : 제거 실습
import pandas as pd
import numpy as np

#결측치 확인
print(df5.isnull().sum())
# isnull() == isna()
print(df5.isna().sum())
#결측치 처리 1 : 결측치가 있는 행 제거
df5_drop_row = df5.dropna()
print(df5_drop_row)
# 결측치 처리 2 : 결측치가 있는 열 제거
df5_drop_col = df5.dropna(axis=1) # dropna() : 특정 라인 제거 parameter axis가 0이면 row, 1이면 columns
print(df5_drop_col)
# 결측치 처리 3 : 결측치를 평균값으로 대체(수에 한해 가능, 문자열인 city는 평균값으로 처리할 수 없음)
df5['age'] = df5['age'].fillna(df5['age'].mean())
print(df5)
# 결측치 처리 4 : 결측치를 최빈값으로 대체
df5['city'] = df5['city'].fillna(df5['city'].mode()[0]) # 가장 많이 나온 값 중 ABC 순으로 정렬되는 듯
print(df5)
# 결측치 처리 5 : 선형 보간법
df5['age'] = df5['age'].interpolate()
print(df5)

# 어떤게 좋다기보다 해당하는 데이터에 따라 사용하는 방법이 달라짐
```

```python
# 데이터프레임 문자열 처리 : 문자열 데이터에서 대문자/소문자 변환, 공백 제거, 특정 문자 대체 방법
import pandas as pd

df = pd.DataFrame({
    'name' : ['Alice ','BOB','CharLie ','dAvE'],
    'city' : ['Seoul','BUSAN ','Daegu ','incheon']
})
print(df)

# 대소문자 변환 : df.str.lower() / df.str.upper()
df['name_lower'] = df['name'].str.lower()
df['name_upper'] = df['name'].str.upper()
# print(df)

# 공백 제거 : df.str.strip()
df['name_strip'] = df['name'].str.strip()
df['city_strip'] = df['city'].str.strip()
# print(df)

# 특정 문자 대체 : df.str.replace()
df['name_replace'] = df['name_strip'].str.replace('a','@',case=False) # case=False : 대소문자 구분하지 않는 것
# print(df)

# 문자열 길이 및 포함 확인
df['name_length'] = df['name_strip'].str.len()
df['has_char_l'] = df['name_strip'].str.contains('l', case=False)
print(df)
```

> 이전 초창기 때의 문제점은 활용하기 어려운 수준의 데이터셋 제공

> 최근 문제점은 해당 데이터에 맞는 공공기관 별로 데이터를 따로 보관

```python
# 데이터프레임 : 데이터 정렬
import pandas as pd

# 예시 데이터
data = {
    'name':['Alice','Bob','Charlie','David'],
    'score':[85,92,78,90],
    'join_date':['2025-03-01','2025-02-15','2025-04-10','2025-01-30']
}
df = pd.DataFrame(data)
# print(df)

# 날짜형으로 변환 : 문자열 데이터를 날짜형 데이터로 변환해야 관련 메소드 사용가능함
df['join_datetime'] = pd.to_datetime(df['join_date'])
# print(df)
# print(df.dtypes)

# 점수 기준 내림차순 정렬
df_sorted = df.sort_values(by='score',ascending=False) # ascending(오름차순) : default=True
print(df_sorted)
```

```python
# 데이터프레임 : 데이터 필터링
high_score = df[df['score']>=85] # score가 85 이상인 데이터만 추출

print(high_score)
print(type(high_score))

date_filterd = df.query("join_date >= '2025-03-01'") # join_date가 2025년 2월 이후인 데이터만 추출
print(date_filterd)
```

```python
# 데이터프레임 : 데이터 합치기 1
# 데이터를 병합하기 위해 pandas.merge() 메소드 활용 : 기준(여기서는 'id)을 가지고 병합
# 학생 정보
students = pd.DataFrame({
    'id': [1,2,3],
    'name':['Alice','Bob','Charlie']
})

# 시험 점수
scores = pd.DataFrame({
    'id':[1,2,4],
    'score':[85,92,78]
})

# ID를 기준으로 inner join : 양 쪽 모두 존재하는 키만 병합
df_merged_inner = pd.merge(students,scores,on='id',how='inner')
print(df_merged_inner)

# ID를 기준으로 outer join : 양 쪽의 키를 모두 포함하고, 없는 값은 NaN으로 채움
df_merged_outer = pd.merge(students,scores,on='id',how='outer')
print(df_merged_outer)
```

```python
# 데이터 프레임 : 데이터 합치기 2
# 데이터를 수직으로 병합하기 위해 concat() 메소드 활용 : 인덱스에 상관없이 통째로 합치기

# 두 데이터프레임 수직으로 결합
df1 = pd.DataFrame({
    'name' : ['Alice','Bob'],
    'score' : [85,90]
})
df2 = pd.DataFrame({
    'name':['Charlie','David'],
    'score':[78,92]
})
print(df1,"\n")
print(df2,"\n")

df_concat = pd.concat([df1,df2],ignore_index=True) # ignore_index : False 설정 시 인덱스가 겹치는 rows 발생 > 기본 True로 두고 사용하는게 일반적임
print(df_concat)
```
