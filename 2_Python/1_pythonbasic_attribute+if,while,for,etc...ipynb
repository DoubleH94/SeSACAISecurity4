```python
# 문자열 리스트 안에 특정 문자 또는 문자열이 포함되어 있는지 체크하는 법
string_list = ['apple', 'banana', 'cherry']
# 1. 정확히 일치하는 문자열이 있는지 확인
if "banana" in string_list :
  print("1.있음")
else :
  print("1,없음")

# 2. 요소 중 일부 문자열(부분 문자열)이 포함되어 있는지 확인
if any("an" in s for s in string_list):
  print("2.부분 문자열 있음")
else:
  print("2.없음")

# 3. 조건에 맞는 모든 문자열 찾기(필터링)
matches = [s for s in string_list if "an" in s]
print("3. ",end="")
for i in range(len(matches)):
    print(matches[i], end="")
    if i != len(matches)-1:
      print(",",end="")
print() # 개행문자 넣지 않아도 기본으로 들어있음

# print(matches)
# print(type(matches))

# 4. 대소문자 무시하고 검사(소문자로 통일)
target = "APPLE"
if any(target.lower() in s.lower() for s in string_list):
  print("4.있음(대소문자 무시)")
```

```python
# if문 예제2 : 점수를 입력 받아 성적을 출력하는 프로그램을 만들자
#(풀이1) : 단순 if문 사용하기
score = int(input('성적을 입력하세요 : '))

if (score >= 0) & (score <= 100) :
  print("당신의 성적은 ",end="")
  if(score >= 90) :
    print("A입니다.")
  elif(score >= 80) :
    print("B입니다.")
  elif(score >= 70) :
    print("C입니다.")
  elif(score >= 60) :
    print("D입니다.")
  else :
    print("F입니다.")
else :
  print("성적입력 오류입니다. 0~100 사이의 점수를 입력했는지 확인해주세요")

# 사람이 이해하기 쉽게 작성하면 실수할 확률이 늘어남( (score >= 90) & (score <=100) )
# 예외처리는 충분히 해줘야한다(추가 : 정수 외에 실수 등으로 입력)
```

```python
# if문 예제2 : 점수를 입력 받아 성적을 출력하는 프로그램을 만들자
#(풀이2) : 딕셔너리를 사용해보자
flag = False
print(type(flag))
score = int(input('성적을 입력하세요 : '))


grade_rules = {
        'A': lambda s: 90 <= s <= 100,
        'B': lambda s: 80 <= s < 90,
        'C': lambda s: 70 <= s < 80,
        'D': lambda s: 60 <= s < 70,
        'F': lambda s: 0 <= s < 60
    }

for grade, rules in grade_rules.items():
    if rules(score):
        print(f"당신의 성적은 {grade}입니다.")
        flag = True
        break
if (flag == False) :
  print("성적입력 오류입니다. 0~100 사이의 점수를 입력했는지 확인해주세요")
#grade = grade_rules(score)  # 딕셔너리를 함수처럼 호출하면 Typeerror 발생, 동작X
```

```python
# self의 역할 : 자기자신의 변수와 매개변수로 받아온 변수를 구분하기위해 사용
class Robot:
  def __init__(self, build_year, name='asdfasd' ):
    self.name = name
    self.build_year = build_year
    self.weight = '45kg'
# 생성자의 매개변수 default 값을 넣기위해서는 ='' 과 같은 양식을 쓰면 된다
# 한 매개변수의 디폴트 값을 설정하면 그 뒤로 나오는 매개변수는 전부 디폴트 값이 있어야한다
# 매개변수의 순서를 바꾸면 디폴트 값을 설정하지 않을 수 있다 > 하지만 이는 인스턴트 생성 시 입력 순서가 바뀌니 주의할 것!!!
robot1 = Robot('pybot', 2045)
robot2 = Robot('thonbot', 2030)
print(robot1.name, robot1.build_year, robot1.weight)
print(robot2.name, robot2.build_year, robot2.weight)
print()
robot2.name = 'thonbot'
robot2.weight = '50kg'
print(robot1.name, robot1.build_year, robot1.weight)
print(robot2.name, robot2.build_year, robot2.weight)

robot3=Robot(2045)
print(robot3.name, robot3.build_year, robot3.weight)
```

```python
# 집합1 : 합집합
A = {1, 2, 3, 4, 5, 6}
B = {1, 3, 4, 7, 8, 9}

print(A | B, '-> A | B')
print(A.union(B),'-> A.union(B)')
print(B.union(A),'-> B.union(A)')
print(A, '-> A')

# 집합2 : 교집합

print(A & B, '-> A & B')
print(A.intersection(B),'-> A.intersection(B)')
print(B.intersection(A),'-> B.intersection(A)')

# 집합3 : 차집합

print(A - B, '-> A - B')
print(A.difference(B),'-> A.difference(B)')
print(B.difference(A),'-> B.difference(A)')

# 집합4 : 대칭차집합(합집합-교집)

print(A ^ B, '-> A ^ B')
print(A.symmetric_difference(B),'-> A.symmetric_difference(B)')

# 집합 5 : 부분집합

print({4, 5, 6} <= A)
print({4, 5, 6}.issubset(A))
print({1, 2, 3, 4, 5, 6} >= A)
print({1, 2, 3, 4, 5, 6}.issuperset(A))
```

```python
# 내장 모듈 예제 > 교수님 코드
# 다음 문장에서 'the'나 'that'과 같은 불용어를 제거하자
sentence = 'The regret after not doing something is bigger than that of doing something'
stop_word=['that','the','of']

def remove_stop_words(sentence):
    words = sentence.split()  # 문자열을 공백을 기준으로 분리하여 리스트로 저장
    new_words=[]
    for word in words:
      if word.lower() not in stop_word :
        new_words.append(word)
    return " ".join(new_words)
print(remove_stop_words(sentence))
```

```python
# 내장모듈 datetime 예제
# 오늘로부터 입력 받은 날이 지난 날은 무슨 요일인지 출력해보자
# ex)
# 날 수를 입력하세요 : 5
# 5일 후는 화요일입니다.
import datetime

d=datetime.datetime.now()
# print(d.weekday())
input_weekday = int(input('날수를 입력하세요:'))
# print(input_weekday)
def whatweekday(input_weekday):
  input_weekday %= 7    # 6이상의 숫자에 대해 요일은 계속 7일주기이기 때문에
  match input_weekday:
    case 0:
      return '월요일'
    case 1:
      return '화요일'
    case 2:
      return '수요일'
    case 3:
      return '목요일'
    case 4:
      return '금요일'
    case 5:
      return '토요일'
    case 6:
      return '일요일'
    case _:
      return 0
print(whatweekday(d.weekday()+input_weekday))

# 교수님 코멘트 : 리스트를 만들어 놓고 인덱스로 접근하면 훨씬 코드가 짧아질
```

```python
# 예외 처리 예제
# 단어를 입력하여 해당 단어가 파일에 존재하지 않으면 추가하는 프로그램을 만들자
# word=input('단어를 입력하세요:')
# print('신규 단어가 추가되었습니다.')
# print('해당 단어가 이미 존재합니다.')

filename = 'words2.txt'

try:
  with open(filename,'r',encoding='utf-8') as file:
    words = [line.strip() for line in file]
except FileNotFoundError:
  words=[]

while(True):
  word = input('단어를 입력하세요:').strip()
  if word == 'q':
      break

  if word in words:
    print('해당 단어가 이미 존재합니다.')
  else:
    words.append(word)
    with open(filename,'a',encoding='utf-8') as file:
      file.write(word+'\n')
    print('신규 단어가 추가되었습니다.')

with open(filename,'r',encoding='utf-8') as file:
  words = [line.strip() for line in file]
print(words)

```

```python
# 단어를 입력하여 해당 단어가 파일에 존재하지 않으면 추가하는 프로그램을 만들자
# 교수님 코드
while(True):
  inputWord = input('단어를 입력하세요:')
  if inputWord == 'q':
    break
  else:
      try:
        with open('dic.txt','a+', encoding='utf-8') as f: # a+ : 읽고 쓰기 동시에(쓰기는 뒤에다가 추가)
          f.seek(0) # a+는 파일포인터가 맨 뒤에 있는데 이걸 맨 앞으로 보냄
          dictionary = f.read() # 파일 전체를 한 번에 읽어옴
          dicList = dictionary.split(' ') # 읽어온 데이터를 단어로 리스트에 저장

          isExist = False # flag

          for item in dicList:
            if inputWord == item:
              isExist == True
              break
          if not isExist:
            f.write((inputWord+' '))
            print('신규 단어가 추가되었습니다.')
          else:
            print('해당 단어가 이미 존재합니다')
      except IOError as err:
        print('에러발생',err)
```
